### literal_eval: literal_eval is a function provided by the ast module in Python. It is used to safely evaluate (parse and interpret) a string containing a Python literal or container display (e.g., dictionaries, lists, tuples) into the corresponding Python object. It can handle various Python data types and structures.

### For example, if 'SECONDARY_TAGS' contains strings like "[1, 2, 3]" or "{'tag1': 'value1', 'tag2': 'value2'}", literal_eval will convert them into the respective Python lists or dictionaries.

### Extraction  purpose of listed propert whether for sales or rent

```python
def extract(description):
    for word in description:
        if 'sale' in word.lower():
            return 'Sale'
        elif 'rent' in word.lower():
            return 'Rent'
    return 'Unknown'
```

```python
## o=for testing purpose

def test(t):
    for word in t:
        if 'sale' in t.lower():
            return 'Sale'
        elif 'rent' in t.lower():
            return 'Rent'
        elif 'flat' in t.lower():
            return 'Flat'
        elif 'furnished' in t.lower():
            return 'Furnished'
```

```python
# Apply the extract function to create a new 'PURPOSE' column
df['PURPOSE'] = df['DESCRIPTION'].str.split(' ').apply(extract)
```
