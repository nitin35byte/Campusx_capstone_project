```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os
import time
# User Agent
headers = { 'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/76.0.3809.132 Safari/537.36'}
```

```python
import time
import os
import pandas as pd
from bs4 import BeautifulSoup
import requests

# Put start page number and end page number
start = 1  # Starting Page
end = 100  # End Page

csv_file = 'Delhi_2.csv'

pageNumber = start
req = 0
while pageNumber < end:
    i = 1
    url = f'https://www.99acres.com/residential-land-in-delhi-ffid-page-{pageNumber}'
    page = requests.get(url, headers=headers)
    pageSoup = BeautifulSoup(page.content, 'html.parser')
    req += 1
    search_div = pageSoup.find('div', {'data-label': 'SEARCH'})
    if search_div:
        sections = search_div.find_all('section', {'data-hydration-on-demand': 'true'})
        for soup in sections:
            # Extract property name and property sub-name
            try:
                property_name = soup.select_one('a.srpTuple__propertyName').text.strip()
                # Extract link
                link = soup.select_one('a.srpTuple__propertyName')['href']
                society = soup.select_one('#srp_tuple_society_heading').text.strip()
            except:
                continue

            # Detail Page
            page = requests.get(link, headers=headers)
            dpageSoup = BeautifulSoup(page.content, 'html.parser')
            req += 1
            try:
                # price Range
                price = dpageSoup.select_one('#pdPrice2').text.strip()
            except:
                price = ''

            # Area
            try:
                area = dpageSoup.select_one('#srp_tuple_secondary_area').text.strip()
            except:
                area = ''

            # Address
            try:
                address = dpageSoup.select_one('#address').text.strip()
            except:
                address = ''

            # Floor Number
            try:
                floorNumAllowed = dpageSoup.select_one('#floorNumLabel').text.strip()
            except:
                floorNumAllowed = ''

            try:
                noOfOpenSides = dpageSoup.select_one('#noOfOpenSides').text.strip()
            except:
                noOfOpenSides = ''

            try:
                possession = dpageSoup.select_one('#agePossessionLbl').text.strip()
            except:
                possession = ''

            # Nearby Locations
            try:
                nearbyLocations = [i.text.strip() for i in
                                   dpageSoup.select('div.NearByLocation__tagWrap span.NearByLocation__infoText')]
            except:
                nearbyLocations = []

            # Descriptions
            try:
                description = dpageSoup.select_one('#description').text.strip()
            except:
                description = ''

            # Features
            try:
                features = [i.text.strip() for i in dpageSoup.select('#features')[0].select('li')]
            except:
                features = []

            # Rating by Features
            try:
                rating = [i.text for i in
                          dpageSoup.select_one('div.review__rightSide>div>ul>li>div div.ratingByFeature__circleWrap')]
            except:
                rating = ''

            try:
                # Property ID
                property_id = dpageSoup.select_one('#Prop_Id').text.strip()
            except:
                property_id = ''

            # Create a dictionary with the given variables
            property_data = {
                'property_name': property_name,
                'link': link,
                'society': society,
                'price': price,
                'areaWithType': area,
                'address': address,
                'floorNumAllowed': floorNumAllowed,
                'noOfOpenSides': noOfOpenSides,
                'possession': possession,
                'nearbyLocations': nearbyLocations,
                'description': description,
                'features': features,
                'rating': rating,
                'property_id': property_id
            }

            temp_df = pd.DataFrame.from_records([property_data])
            i += 1
            if os.path.isfile(csv_file):
                # Append DataFrame to the existing file without header
                temp_df.to_csv(csv_file, mode='a', header=False, index=False)
            else:
                # Write DataFrame to the file with header
                temp_df.to_csv(csv_file, mode='a', header=True, index=False)

    print(f'{pageNumber} -> {i}')
    pageNumber += 1

    if req % 4 == 0:
        time.sleep(10)
    if req % 20 == 0:
        time.sleep(20)
```
